```python
from sklearn.model_selection import train_test_split

# we add stratify to ensure the 16% imbalance is preserved in all splits

df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=1, stratify=df['Attrition'])
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=1, stratify=df_full_train['Attrition'])
```

```python
X_train=df_train.reset_index(drop=True)
X_val=df_val.reset_index(drop=True)
X_test=df_test.reset_index(drop=True)
X_full_train=df_full_train.reset_index(drop=True)
X_train
```

```python
y_train = df_train.Attrition
y_val = df_val.Attrition
y_test = df_test.Attrition
y_full_train = df_full_train.Attrition
```

```python
del X_train['Attrition']
del X_val['Attrition']
del X_test['Attrition']
del X_full_train['Attrition']
```

```python
corr_with_target = X_train[numerical].corrwith(y_train)
corr_with_target
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical),
        ('cat', categorical_transformer, categorical)
    ],
)

preprocessor
```

```python
X_train_processed = preprocessor.fit_transform(X_train)
X_val_processed = preprocessor.transform(X_val)
X_test_processed = preprocessor.transform(X_test)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
# class_weight='balanced'  automatically fixes the imbalance issue we saw in the histogram.
clf = LogisticRegression(solver='liblinear', class_weight='balanced', random_state=1)

clf.fit(X_train_processed, y_train)

y_pred_val = clf.predict_proba(X_val_processed)[:, 1]
y_pred_train = clf.predict_proba(X_train_processed)[:, 1]
print(f"Logistic Regression AUC: {roc_auc_score(y_val, y_pred_val):.4f}")
print(f"Logistic Regression (train) AUC: {roc_auc_score(y_train, y_pred_train):.4f}")
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(
        n_estimators=100,     
        max_depth=1,          
        random_state=1,       
        n_jobs=-1,            
        class_weight='balanced' 
    )
rf.fit(X_train_processed, y_train)
    
y_pred_rf = rf.predict_proba(X_val_processed)[:, 1]
auc = roc_auc_score(y_val, y_pred_rf)
y_train_rf = rf.predict_proba(X_train_processed)[:, 1]
auc_train = roc_auc_score(y_train, y_train_rf)
    
print(f"val -> AUC: {auc:.4f}")
print(f"train -> AUC: {auc_train:.4f}")
```

```python
# We loop through different depths to see which one works best
depths = [1,2, 3, 5 , 7 ,10, 15, 20, 25]
scores = []
for d in depths:
    rf = RandomForestClassifier(
        n_estimators=100,     
        max_depth=d,          
        random_state=1,       
        n_jobs=-1,            
        class_weight='balanced' 
    )
    rf.fit(X_train_processed, y_train)
    
    y_pred_rf = rf.predict_proba(X_val_processed)[:, 1]
    auc = roc_auc_score(y_val, y_pred_rf)

    y_train_rf = rf.predict_proba(X_train_processed)[:, 1]
    auc_train = roc_auc_score(y_train, y_train_rf)
    
    print(f"Depth: {d} ")
    print(f"val -> AUC: {auc:.4f}")
    print(f"train -> AUC: {auc_train:.4f}")
    print("-----------------------------")
    scores.append((d, auc))


best_depth, best_score = max(scores, key=lambda x: x[1])
print(f"\nBest Random Forest AUC: {best_score:.4f} (Depth {best_depth})")
```

```python
# We loop through different depths to see which one works best
n_estimators = [50,100, 200, 300 , 400 , 500]
scores = []
for n in n_estimators:
    rf = RandomForestClassifier(
        n_estimators=n,     
        max_depth= 3,          
        random_state=1,       
        n_jobs=-1,            
        class_weight='balanced' 
    )
    rf.fit(X_train_processed, y_train)
    
    y_pred_rf = rf.predict_proba(X_val_processed)[:, 1]
    auc = roc_auc_score(y_val, y_pred_rf)

    y_train_rf = rf.predict_proba(X_train_processed)[:, 1]
    auc_train = roc_auc_score(y_train, y_train_rf)
    
    print(f"n_estimators: {n} ")
    print(f"val -> AUC: {auc:.4f}")
    print(f"train -> AUC: {auc_train:.4f}")
    print("-----------------------------")
    scores.append((n, auc))


best_n_estimators, best_score = max(scores, key=lambda x: x[1])
print(f"\nBest Random Forest AUC: {best_score:.4f} (n_estimators {best_n_estimators})")
```

```python
leaf_options = [1, 3, 5, 10, 15]
depths = [ 3, 7 ,10, 15]
for d in depths:
    print(f"(with max_depth={d})...")
    for leaf in leaf_options:
        rf = RandomForestClassifier(
            n_estimators=100,
            max_depth=d,         
            min_samples_leaf=leaf, 
            random_state=1,
            n_jobs=-1,
            class_weight='balanced'
        )
        
        rf.fit(X_train_processed, y_train)
        
        y_pred_train = rf.predict_proba(X_train_processed)[:, 1]
        y_pred_val = rf.predict_proba(X_val_processed)[:, 1]
        
        auc_train = roc_auc_score(y_train, y_pred_train)
        auc_val = roc_auc_score(y_val, y_pred_val)
        print("-----------------------------")
        print(f"Min Samples Leaf: {leaf}")
        print(f"  Train AUC: {auc_train:.4f}")
        print(f"  Val AUC:   {auc_val:.4f}")
        
    print("-" * 40)
```

best result `max_depth=15` with `min_samples_leaf=3`

#### **Logistic Regression:** `0.8046` (Very simple model, doesn't overfit).

#### **Random Forest**: `0.8045` (Huge model, Train AUC is 0.9999 so it is still memorizing, but generalizing decently).

# XGBoost

```python
import xgboost as xgb  
```

```python
features = preprocessor.get_feature_names_out().tolist()

dtrain = xgb.DMatrix(X_train_processed, label=y_train, feature_names=features)
dval = xgb.DMatrix(X_val_processed, label=y_val, feature_names=features)

watchlist = [(dtrain, 'train'), (dval, 'val')]

xgb_params = {
    'eta': 0.1,
    'max_depth': 6,
    'min_child_weight': 1,
    'objective': 'binary:logistic',
    'eval_metric': 'auc',
    'scale_pos_weight': 5,
    'nthread': 8,
    'seed': 1,
    'verbosity': 1
}

model_xgb = xgb.train(xgb_params, dtrain, num_boost_round=100, evals=watchlist, verbose_eval=10)

y_pred_xgb = model_xgb.predict(dval)
xgb_score = roc_auc_score(y_val, y_pred_xgb)

print(f"\n--------------------------------")
print(f"XGBoost:             {xgb_score:.4f}")
print(f"--------------------------------")
```

```python
features = preprocessor.get_feature_names_out().tolist()
dtrain = xgb.DMatrix(X_train_processed, label=y_train, feature_names=features)
dval = xgb.DMatrix(X_val_processed, label=y_val, feature_names=features)


etas = [0.3, 0.1, 0.05, 0.01]  
depths = [3, 6, 10]           


best_xgb_score = 0
best_params = {}

for eta in etas:
    for depth in depths:

        xgb_params = {
            'eta': eta,
            'max_depth': depth,
            'min_child_weight': 1,
            'objective': 'binary:logistic',
            'eval_metric': 'auc',
            'scale_pos_weight': 5,  
            'nthread': 8,
            'seed': 1,
            'verbosity': 0
        }
        
        model_xgb = xgb.train(xgb_params, dtrain, num_boost_round=100, evals=watchlist, verbose_eval=10)
        
        y_pred_xgb = model_xgb.predict(dval)
        xgb_score = roc_auc_score(y_val, y_pred_xgb)
        
        print(f"eta={eta}, depth={depth} -> AUC: {xgb_score:.4f}")
        print("-" * 40)
        if xgb_score > best_xgb_score:
            best_xgb_score = xgb_score
            best_params = xgb_params

print("-" * 40)
print(f"Best XGBoost AUC: {best_xgb_score:.4f}")
print(f"Best Params: {best_params}")
```

```python
features = preprocessor.get_feature_names_out().tolist()

dtrain = xgb.DMatrix(X_train_processed, label=y_train, feature_names=features)
dval = xgb.DMatrix(X_val_processed, label=y_val, feature_names=features)

watchlist = [(dtrain, 'train'), (dval, 'val')]

xgb_params = {
    "max_depth": 3,
    "eta": 0.3,
    "min_child_weight": 5,
    "gamma": 0.2,
    "subsample": 0.8,
    "colsample_bytree": 0.8,
    'objective': 'binary:logistic',
    'eval_metric': 'auc',
    'scale_pos_weight': 5,
    'nthread': 8,
    'seed': 1,
    'verbosity': 1
}

model_xgb= xgb.train(
    xgb_params,
    dtrain,
    num_boost_round=500,
    evals=watchlist,
    early_stopping_rounds=10
)

y_pred_xgb = model_xgb.predict(dval)
xgb_score = roc_auc_score(y_val, y_pred_xgb)

print(f"\n--------------------------------")
print(f"XGBoost:             {xgb_score:.4f}")
print(f"--------------------------------")
```

```python
xgb_cv = xgb.cv(
    xgb_params,
    dtrain,
    num_boost_round=300,
    nfold=5,
    early_stopping_rounds=15,
    metrics="auc",
    seed=1
)

model_xgb = xgb.train(
    xgb_params,
    dtrain,
    num_boost_round=500,
    evals=watchlist,
    early_stopping_rounds=15,
    verbose_eval=False
)

y_pred_xgb = model_xgb.predict(
    dval,
    iteration_range=(0, model_xgb.best_iteration + 1)
)

xgb_score = roc_auc_score(y_val, y_pred_xgb)

print("\n--------------------------------")
print(f"XGBoost AUC: {xgb_score:.4f}")
print("--------------------------------")
```

#### **Logistic Regression:** `0.8046` (Very simple model, doesn't overfit).

#### **Random Forest**: `0.8045` (Huge model, Train AUC is 0.9999 so it is still memorizing, but generalizing decently).
#### **XGBoost**: `0.8208` (Huge model, Train AUC is 0.99424 so it is still memorizing, but generalizing decently).

# Training the final model

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

pipeline = Pipeline(steps=[
    
    ('preprocessor', preprocessor), 
    
    ('model',  LogisticRegression(
        C=0.01,   
        random_state=1,
        solver="liblinear",
        class_weight="balanced"
    ))
])
```

```python
pipeline.fit(X_full_train, y_full_train)
```

```python
# saving the model 
import pickle
with open('model.bin', 'wb') as f_out:
    pickle.dump(pipeline, f_out)
```
